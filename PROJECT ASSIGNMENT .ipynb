```python
#Importing the necessary library


from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
url="https://obiwezy.com/computing/laptops.html"  # getting the url
page=requests.get(url)
sp=BeautifulSoup(page.text,'html.parser')
```

```python
product_name = sp.find_all('a', href=True, class_='product-item-link')
prices = sp.find_all('span', class_='price')

```

```python
data = [] # declaring a list to hold the product_name and price
for name, price in zip(product_name, prices):
    data.append({
        'Product Name': name.text,
        'Price': price.text,
        
    })
```

```python
#Slicing the Product name to get the brand Names
sliced_product_names = [item['Product Name'].split(',')[0] for item in data]

```

```python
sliced_product_names
```

```python
# Converting The data to a Pandas Table
Data_table=pd.DataFrame(data)

```

```python
Data_table
```

```python
# Updating the product Name in the Data Table
if len(sliced_product_names) == len(Data_table):
    Data_table['Product Name'] = sliced_product_names
else:
    print("The length of sliced_product_names does not match the length of Data_table.")
```

```python
Data_table
```

```python
#Checking the the values of the Price column 

print(Data_table['Price'].unique())
```

```python
# removing the NGN and the \xao from the price column

Data_table['Price'] = Data_table['Price'].str.replace('\xa0', '').str.replace(',', '').str.replace('NGN','').astype(str)
```

```python
print(Data_table['Price'].unique())
```

```python
#creating Visualization form the estracted data in the Data_table
```

```python
# Defining  custom colors and bar width
custom_colors = ['#03045e', '#023e8a', '#0077b6', '#0096c7', '#00b4d8','#48cae4', '#90e0ef', '#ade8f4','#caf0f8']
bar_width = 0.5  

# Create a figure and axis
plt.figure(figsize=(12, 6)) 
ax = plt.gca()

# Extract data displaying the top ten product in the Data_table
product_names = Data_table['Product Name'].head(10)
prices = Data_table['Price'].head(10)

# Create bar chart
bars = plt.bar(product_names, prices, width=bar_width, color=custom_colors)

# Customize labels and title
plt.xlabel('Product Name')
plt.ylabel('Prices -NGN')
plt.title('Product against Prices')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
